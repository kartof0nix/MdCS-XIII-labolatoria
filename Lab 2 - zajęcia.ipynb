```python


def calculate_values(base_array, func):
  result = []
  for i in base_array:
    result.append(func(i))
  return result

def arrange(st, end, tick):
  res = []
  while(st <= end):
    res.append(st)
    st+= tick
  return res
```

```python
def t(x):
  return x**2
def y(x):
  return 2**x
  
base_y = list(range(1, 10))
result_y = calculate_values(base_y, y)

base_t = list(range(1, 10))
result_t = calculate_values(base_t, t)

print(base_y, result_y)
```

```python
plt.plot(base_y, result_y)
plt.plot(base_t, result_t)



plt.ylabel("y")
plt.xlabel("x")
plt.show()
```
